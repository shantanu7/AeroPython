## Code to plot two test functions:

```python
%matplotlib inline
import numpy
import matplotlib
from matplotlib import pyplot
```

```python
#Plot y = x^3

x1 = numpy.linspace(-4,4);
y1 = x1**3;
```

```python
pyplot.plot(x1,y1);
pyplot.grid(True)
pyplot.xlabel('$x$')
pyplot.ylabel('$y$')
pyplot.title('Graph of $y = x^{3}$');
```

```python
#Test graph for potential flow around a corner

ngx = 101; ngy = 101;
A = 1.0 #Parameter for potential flow

x = numpy.linspace(0.0, 2.0, ngx);
y = numpy.linspace(0.0, 2.0, ngy);

X, Y = numpy.meshgrid(x,y);

u = numpy.zeros((ngx, ngy));
v = numpy.zeros_like(u);

u = 2.0*A*X; v = -2.0*A*Y;
```

```python
pyplot.figure(figsize=(6,6))
pyplot.streamplot(X, Y, u, v);
pyplot.xlim(0,2.0);
pyplot.ylim(0,2.0);
pyplot.xlabel('$x$')
pyplot.ylabel('$y$')
pyplot.title('Streamlines for potential flow around a corner');
```
