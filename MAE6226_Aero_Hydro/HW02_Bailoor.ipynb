```python
import math
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
#Define grid parameters

N = 100
Xmin = -5.0; Xmax = 5.0
Ymin = -5.0; Ymax = 5.0

x = numpy.linspace(Xmin, Xmax, N);
y = numpy.linspace(Ymin, Ymax, N);

dx = x[1]-x[0]; dy = y[1]-y[0];

X, Y = numpy.meshgrid(x, y)
```

```python
#Define flow parameters

sigma_source = 2.0;               #Source strength
x_source, y_source = -2.0, 0.0

sigma_sink = -2.0;               #Sink strength
x_sink, y_sink = 2.0, 0.0

u_source = sigma_source/(2.*math.pi)*(X-x_source)/((X-x_source)**2+(Y-y_source)**2)
v_source = sigma_source/(2.*math.pi)*(Y-y_source)/((X-x_source)**2+(Y-y_source)**2)

u_sink = sigma_sink/(2.*math.pi)*(X-x_sink)/((X-x_sink)**2+(Y-y_sink)**2)
v_sink = sigma_sink/(2.*math.pi)*(Y-y_sink)/((X-x_sink)**2+(Y-y_sink)**2)

xvelo = u_source + u_sink
yvelo = v_source + v_sink
Velocity = numpy.sqrt(u_source**2 + v_source**2) + numpy.sqrt(u_sink**2 + v_sink**2)
```

```python
pyplot.figure(figsize=(15, 6))

pyplot.subplot(121)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(Xmin, Xmax)
pyplot.ylim(Ymin, Ymax)
pyplot.streamplot(X, Y, xvelo, yvelo, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
pyplot.scatter([x_source, x_sink], [y_source, y_sink], color='#CD2305', s=80, marker='o', linewidth=0);
pyplot.title('Fig. 1(a): Streamlines for $\sigma_{source} = 2.0, \sigma_{sink} = -2.0 $')

pyplot.subplot(122)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(Xmin, Xmax)
pyplot.ylim(Ymin, Ymax)
pyplot.contour(X,Y, Velocity, levels = numpy.linspace(0,4,21))
pyplot.title('Fig. 1(b): Velocity potential for $\sigma_{source} = 2.0, \sigma_{sink} = 2.0 $')
```

<b><u>Explanation P1:</u></b> <br>

When the strengths of the source and sink are equal, all flow leaving the source enters the sink. The streamlines are symmetrical about the midplane of the source-sink couple, as well as the line passing through the source and sink.

When the strength of the source is greater than that of the sink, the source emits more flow than can be consumed by the sink. Thus, a part of the flow emitted by the source exits the domain, as seen in Fig. 3(a) ($\sigma_{source} = 2.0$, $\sigma_{sink} = -0.5$). The fraction of flow entering the sink is proportional to the ratio $\frac{\sigma_{sink}}{\sigma_{source}}$. Thus, in case (a), all flow leaving the source enters the sink, in case (b) half the flow leaving the source enters the sink and in case (c) a quarter of the flow leaving the source enters the sink.

Also, the flow is divided into two parts: one that enters the sink and the other that leaves the domain. The streamline demarcating the two parts is known as the dividing streamline. Slightly downstream of the sink, a stagnation point is also observed. This stagnation point is not seen in the case in which $\sigma_{sink}$ = $\sigma_{source}$. Streamlines are symmetrical about the line passing through the source and sink but not along the normal line passing through the midpoint of the source-sink pair.

```python
#Define flow parameters

sigma_source = 1.0;               #Source strength
x_source, y_source = -0.0, 0.0

u_source = sigma_source/(2.*math.pi)*(X-x_source)/((X-x_source)**2+(Y-y_source)**2)
v_source = sigma_source/(2.*math.pi)*(Y-y_source)/((X-x_source)**2+(Y-y_source)**2)

xvelo = u_source
yvelo = v_source
Velocity = numpy.sqrt(u_source**2 + v_source**2)
```

```python
size = 8
pyplot.figure(figsize=(size, size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(Xmin, Xmax)
pyplot.ylim(Ymin, Ymax)
pyplot.streamplot(X, Y, xvelo, yvelo, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
pyplot.scatter(x_source, y_source, color='#CD2305', s=80, marker='o', linewidth=0);
pyplot.title('Fig. 4: Streamlines for $\sigma_{source} = 1.0 $')
```

```python
center = 49

#Integrate top and bottom faces
for i in range (2):
    interval = 10 + i*10;
    integral = 0.0
    for jstep in range (2):
        j = center-interval + 2*jstep*interval
        for i in range (center-interval,center+interval):
            integral = integral + 0.5*abs(v_source[i,j]+v_source[i+1,j])*dx

    for istep in range (2):
        i = center-interval + 2*istep*interval
        for j in range (center-interval,center+interval):
            integral = integral + 0.5*abs(u_source[i,j]+u_source[i,j+1])*dy
    
    edge = x[center+interval]-x[center-interval];
    print("")
    print("Integral using square of edge", edge, ":", integral, " units")
```

<b><u>Explanation P2:</u></b> <br>

It is observed that the integral indicating the volume flow rate through two square blocks around the source are considerably smaller than the expected value of 1.0. However, it is slightly larger for the larger control volume. Let's now look at the value of this integral over the entire domain:

```python
integral = 0.0
for i in range (N-1):
    integral = integral + 0.5*abs(v_source[i,0]+v_source[i+1,0])*dx
    integral = integral + 0.5*abs(v_source[i,-1]+v_source[i+1,-1])*dx
    
for j in range (N-1):
    integral = integral + 0.5*abs(v_source[0,j]+v_source[0,j+1])*dx
    integral = integral + 0.5*abs(v_source[-1,j]+v_source[-1,j+1])*dx
    
print("Integral:", integral, " units")
```

<b><u>Explanation P2 (contd):</u></b> <br>

The value of the integral is now closer to the expected value. Neglecting roundoff error, the deviation of the computed values from the expected value can be attributed to discretization errors. As the control volume is made increasingly large, keeping $\Delta x$ constant, the relative discretization error becomes smaller. Also, more sampling points are available for computing the numerical integral. Thus, the value of the integral becomes more accurate as the size of the control volume increases.

```python
#Define coordinates for source/ sink
xso = [-1.0, 0.0, 1.0]; yso = [1.0, 1.0, 1.0];

xsi = [-1.0, 0.0, 1.0]; ysi = [-1.0, -1.0, -1.0];

sigma = 1.0 #Common strength for source/ sink

Velocity = numpy.zeros_like(X);
velx = numpy.zeros_like(X); vely = numpy.zeros_like(X);

for iso in range (3):
    velx = velx +  sigma/(2.*math.pi)* ( (X-xso[iso])/((X-xso[iso])**2+(Y-yso[iso])**2) \
                                       - (X-xsi[iso])/((X-xsi[iso])**2+(Y-ysi[iso])**2) )
    vely = vely +  sigma/(2.*math.pi)* ( (Y-yso[iso])/((X-xso[iso])**2+(Y-yso[iso])**2) \
                                       - (Y-ysi[iso])/((X-xsi[iso])**2+(Y-ysi[iso])**2) ) 

Velocity = numpy.sqrt(velx**2 + vely**2)
xscatter = [xso, xsi]; yscatter = [yso, ysi];
```

```python
size = 7
pyplot.figure(figsize=(size, size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(Xmin, Xmax)
pyplot.ylim(Ymin, Ymax)
pyplot.streamplot(X, Y, velx, vely, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
pyplot.scatter(xscatter, yscatter, color='#CD2305', s=80, marker='o', linewidth=0);
pyplot.title('Fig. 5: Streamlines for three source-sink pairs with $\sigma_{source} = 1.0, \sigma_{sink} = -1.0 $')
```

```python
integral = 0.0

#Integrate top and bottom faces
for jstep in range (2):
    normy = -1.0**(jstep + 1)
    j = 0 + jstep*(N-1)
    for i in range (N-1):
        integral = integral + 0.5*(vely[i,j]+vely[i+1,j])*normy*dx

for istep in range (2):
    normx = -1.0**(istep + 1)
    i = 0 + istep*(N-1)
    for j in range (N-1):
        integral = integral + 0.5*(velx[i,j]+velx[i,j+1])*normx*dy
                
print('Net volumetric flux in domain:', integral, " units")
```

<b><u>Explanation P3:</u></b> <br>
    
Since the sources and sinks are all of the same strength, one might expect the net volumetric flux to be zero. Integrating over the domain, we get a small value (~0.38) and the deviation may be, once again, attributed to discretization errors. This can be tested by using a smaller control volume. A larger value would be then obtained.
